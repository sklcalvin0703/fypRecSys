```python
import os
import csv
import sys
import re
import pandas as pd

from surprise import Dataset
from surprise import Reader
from surprise import SVD
from sklearn.externals import joblib
```

```python
def loadDataSet():
    dir = os.getcwd()

    ratingsPath = dir + '/ml-latest-small/ratings.csv'
    moviesPath = dir + '/ml-latest-small/movies.csv'

    data = pd.read_csv(ratingsPath,header=None,skiprows=1, 
                 names=['userId', 'movieId', 'rating', 'timestamp'])


    processedData = data.drop(['timestamp'],axis=1)

    return processedData
```

```python
data = loadDataSet()
data.head()
```

```python
#save the dataset
data.to_pickle("training_data.pkl")
```

```python
reader = Reader()
processedData = Dataset.load_from_df(data[['userId', 'movieId', 'rating']], reader)
processedData = processedData.build_full_trainset()
```

```python
print('training in progress:')
```

```python
model = SVD(n_epochs=10, verbose=True)
model.fit(processedData)
joblib.dump(model, "SVD.pkl")
```

```python
#prediction
recommendation = []
the_iid_list = processedData.all_items()
for iid in the_iid_list:
    prediction = model.predict(196, iid)
    intMovieId = int(prediction[1])
    estimatedRating = prediction[3]
    recommendation.append((intMovieId,estimatedRating))
```

```python
print(recommendation)
```
